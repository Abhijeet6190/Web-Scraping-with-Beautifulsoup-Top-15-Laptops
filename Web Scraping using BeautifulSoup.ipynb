```python
# create list
products=[] # list to store names of Laptops
prices=[] # list to store laptop prices
ratings=[] # list to store ratings
```

```python
for data in soup.find_all('a',class_='_1fQZEK'):
    name=data.find('div',attrs={'class':'_4rR01T'})
    price=data.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rating=data.find('div',attrs={'class':'_3LWZlK'})
    
    products.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text) if type(rating)==bs4.element.Tag else ratings.append('NaN')
    
```

If ratings for the product is not available then it will show us Nan value

bs4.element.tag is type

```python
#Creating dataframe

import pandas as pd
df=pd.DataFrame({'Product Name':products,'Price':prices,'Ratings':ratings})
```

```python
df.info()
```

```python
# Format Price Column to remove symbol and ,
df['Price']=df['Price'].str.lstrip('₹')
```

```python
df
```

```python
df['Price']=df['Price'].replace({',':''},regex=True)
```

```python
df['Price']=df['Price'].astype(float)
```

```python
# Save this table in csv
df.to_csv('webscrapping_data.csv',index=False)
```
